```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/ny-flights(1).csv'
ny_flights = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
ny_flights.head(), ny_flights.info()
```

```python
# Question: What is the average departure delay for each airline?

# Calculate the average departure delay for each airline
average_dep_delay_per_airline = ny_flights.groupby('unique_carrier')['dep_delay'].mean().sort_values(ascending=False)
average_dep_delay_per_airline
```

```python
# Question: What is the average arrival delay for each destination airport?

# Calculate the average arrival delay for each destination airport
average_arr_delay_per_dest = ny_flights.groupby('dest')['arr_delay'].mean().sort_values(ascending=False)
average_arr_delay_per_dest
```

```python
# Convert 'fl_date' to datetime format for better date handling
ny_flights['fl_date'] = pd.to_datetime(ny_flights['fl_date'])

# Extract month from the 'fl_date' column
ny_flights['month'] = ny_flights['fl_date'].dt.month

# Question: How does the departure delay vary by month?

# Calculate the average departure delay for each month
average_dep_delay_per_month = ny_flights.groupby('month')['dep_delay'].mean()
average_dep_delay_per_month
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
file_path = '/ny-flights(1).csv'
flights_data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(flights_data.head())
```

```python
# Summary statistics
summary_stats = flights_data.describe()
print(summary_stats)
```

```python

# Histograms
flights_data.hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of NY Flights Dataset Variables')
plt.show()
```

```python
# Box plots
plt.figure(figsize=(15, 10))
for i, column in enumerate(flights_data.columns, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(y=flights_data[column])
    plt.title(f'Box plot of {column}')
plt.tight_layout()
plt.show()
```

```python
# Scatter plots
sns.pairplot(flights_data)
plt.suptitle('Scatter plots of NY Flights Dataset Variables')
plt.show()
```

```python
# Example datasets
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                    'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2']})

df2 = pd.DataFrame({'key': ['K0', 'K1', 'K3'],
                    'C': ['C0', 'C1', 'C3'],
                    'D': ['D0', 'D1', 'D3']})

# Merge based on 'key' column
merged_df = pd.merge(df1, df2, on='key', how='inner')
print(merged_df)
```

```python
# Example datasets
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                    index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C1', 'C2'],
                      'D': ['D0', 'D1', 'D2']},
                     index=['K0', 'K2', 'K3'])

# Join based on index
joined_df = left.join(right, how='inner')
print(joined_df)
```
