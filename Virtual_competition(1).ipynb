**16.** The 'priority' column contains the values 'yes' and 'no'. Replace this column with a column of boolean values: 'yes' should be `True` and 'no' should be `False`.

````python
print(df.columns)
import pandas as pd

# Example DataFrame
data = {
    'animal': ['cat', 'dog', 'cat', 'dog', 'cat', 'rabbit', 'rabbit'],
    'age': [2.5, 3, 1.5, 4, 2, 2, 3],
    'visits': [1, 3, 2, 1, 3, 2, 1],
    'priority': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'no']
}

df = pd.DataFrame(data)

# Replace 'priority' column with boolean values
df['priority'] = df['priority'].map({'yes': True, 'no': False})

print("DataFrame with boolean 'priority' column:")
print(df)

````

**19.** Which airline ID is present maximum times in the dataset

````python
pip install nycflights13
````

````python
from nycflights13 import flights

# Display the first few rows of the dataset
print(flights.head())
````

````python
# Calculate frequency of each airline ID
airline_counts = flights['carrier'].value_counts()

# Find the airline ID with maximum frequency
max_airline_id = airline_counts.idxmax()

# Find the maximum count
max_count = airline_counts.max()

print(f"The airline ID '{max_airline_id}' is present {max_count} times in the dataset.")
````

**20.** Draw a plot between dep_delay and arr_delay

````python
from nycflights13 import flights

# Display the first few rows of the dataset
print(flights.head())
````

````python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(flights['dep_delay'], flights['arr_delay'], alpha=0.5)
plt.title('Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True)
plt.show()
````
