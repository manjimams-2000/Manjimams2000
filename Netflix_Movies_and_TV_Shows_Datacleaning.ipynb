```python
import zipfile
import os
```

```python
zip_file_path = '/content/netflix_titles.csv.zip'
extracted_folder = '/mnt/data/netflix_titles/'
```

```python
os.makedirs(extracted_folder, exist_ok=True)
```

```python
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder)
```

```python
extracted_files = os.listdir(extracted_folder)
extracted_files
```

```python
import pandas as pd
```

```python
csv_file_path = os.path.join(extracted_folder, 'netflix_titles.csv')
df = pd.read_csv(csv_file_path)
```

```python
df.head()
```

```python
df_cleaned = df.dropna(subset=['title', 'type'])
df
```

```python
df_cleaned['director'] = df_cleaned['director'].fillna('Unknown')
df
```

```python
df_cleaned['cast'] = df_cleaned['cast'].fillna('Unknown')
df
```

```python
df_cleaned['country'] = df_cleaned['country'].fillna('Unknown')
df
```

```python
df_cleaned['rating'] = df_cleaned['rating'].fillna('Unknown')
df
```

```python
missing_values = df_cleaned.isnull().sum()
missing_values
```

```python
df_cleaned['date_added'] = df_cleaned['date_added'].fillna('1900-01-01')
df
```

```python
df_cleaned['duration'] = df_cleaned['duration'].fillna('Unknown')
df
```

```python
df_cleaned['date_added'] = pd.to_datetime(df_cleaned['date_added'], format='%Y-%m-%d', errors='coerce')
# Specify the format for the dates you filled in and handle errors
df
```

```python
df_cleaned = df_cleaned.drop_duplicates()
df
```

```python
df_cleaned.head()
df
```

```python
df_cleaned['duration_minutes'] = df_cleaned['duration'].apply(lambda x: int(x.split()[0]) if 'min' in x else 0)
df
```

```python
df_cleaned['num_seasons'] = df_cleaned['duration'].apply(lambda x: int(x.split()[0]) if 'Season' in x else 0)
df
```

```python
df_cleaned = df_cleaned.drop(columns=['duration'])
df
```

```python
print(df_cleaned[['title', 'duration_minutes', 'num_seasons']].head())
df
```

```python
df_cleaned = df_cleaned.drop_duplicates()
print("DataFrame shape after removing duplicates:", df_cleaned.shape)
```

```python
df_cleaned['title'] = df_cleaned['title'].str.lower()
df
```

```python
df_cleaned['director'] = df_cleaned['director'].str.lower()
df
```

```python
df_cleaned['cast'] = df_cleaned['cast'].str.lower()
df
```

```python
df_cleaned['country'] = df_cleaned['country'].str.lower()
df
```

```python
Q1 = df_cleaned['release_year'].quantile(0.25)
Q3 = df_cleaned['release_year'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_cleaned = df_cleaned[(df_cleaned['release_year'] >= lower_bound) & (df_cleaned['release_year'] <= upper_bound)]
df
```

```python
df_cleaned.to_csv('cleaned_dataset.csv', index=False)
print("Cleaned DataFrame saved to 'cleaned_dataset.csv'")
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='type')
plt.title('Count of Movies vs TV Shows')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['release_year'], bins=20, kde=True)
plt.title('Distribution of Release Years')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='release_year')
plt.title('Boxplot of Release Years')
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
```
