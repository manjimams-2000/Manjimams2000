```python
import pandas as pd


```

```python
# Load the dataset
file_path = '/content/House_Pricing.csv'
house_pricing_data = pd.read_csv(file_path)


```

```python
# Display the first few rows of the dataset
house_pricing_data.head()
```

```python
# Check for missing values in the dataset
missing_values = house_pricing_data.isnull().sum()
print(missing_values)
```

```python
numerical_columns = ['Sale Price', 'No of Bedrooms', 'No of Bathrooms', 'Flat Area (in Sqft)',
                     'Lot Area (in Sqft)', 'No of Floors', 'Overall Grade', 'Area of the House from Basement (in Sqft)',
                     'Basement Area (in Sqft)', 'Age of House (in Years)', 'Renovated Year', 'Zipcode',
                     'Latitude', 'Longitude', 'Living Area after Renovation (in Sqft)',
                     'Lot Area after Renovation (in Sqft)']
```

```python
non_numerical_columns = ['ID', 'Date House was Sold', 'Waterfront View', 'No of Times Visited', 'Condition of the House']
```

```python
house_pricing_data[numerical_columns] = house_pricing_data[numerical_columns].fillna(house_pricing_data[numerical_columns].median())
```

```python
house_pricing_data['Date House was Sold'] = house_pricing_data['Date House was Sold'].fillna(house_pricing_data['Date House was Sold'].mode()[0])
house_pricing_data['Waterfront View'] = house_pricing_data['Waterfront View'].fillna('No')
house_pricing_data['No of Times Visited'] = house_pricing_data['No of Times Visited'].fillna('None')
house_pricing_data['Condition of the House'] = house_pricing_data['Condition of the House'].fillna('Good')
```

```python
house_pricing_data[numerical_columns] = house_pricing_data[numerical_columns].apply(pd.to_numeric, errors='coerce')
```

```python
categorical_columns = ['Waterfront View', 'No of Times Visited', 'Condition of the House']
house_pricing_data[categorical_columns] = house_pricing_data[categorical_columns].astype('category')
```

```python
house_pricing_data['Date House was Sold'] = pd.to_datetime(house_pricing_data['Date House was Sold'], errors='coerce')
```

```python
house_pricing_data['Year Sold'] = house_pricing_data['Date House was Sold'].dt.year
house_pricing_data['Month Sold'] = house_pricing_data['Date House was Sold'].dt.month
house_pricing_data['Day Sold'] = house_pricing_data['Date House was Sold'].dt.day
```

```python
house_pricing_data.drop(columns=['Date House was Sold'], inplace=True)
```

```python
house_pricing_data = pd.get_dummies(house_pricing_data, columns=categorical_columns, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
house_pricing_data[numerical_columns] = scaler.fit_transform(house_pricing_data[numerical_columns])
```
